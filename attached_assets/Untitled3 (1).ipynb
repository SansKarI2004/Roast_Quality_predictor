```python
X=df.drop(columns=df.columns[-1])
y=df.iloc[:,-1]
```

```python
X['date_time']=X['date_time'].astype('category')
X['AH_data']=X['AH_data'].astype(float)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2,random_state=32)
```

```python
X_train.shape
```

```python
X_val.shape
```

```python
from xgboost import XGBRegressor

model= XGBRegressor(enable_categorical=True).fit(X_train,y_train)
y_pred=model.predict(X_val)
y_pred_tr=model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
print(f'{mean_squared_error(y_pred,y_val):.2f}')
print(f'{r2_score(y_pred,y_val):.2f}')
print(f'{mean_absolute_error(y_pred,y_val):.2f}')
print(f'{mean_absolute_error(y_pred_tr,y_train):.2f}')
```
